### 7- Write a function which calculate solution set of quadratic equation, solve_quadratic_eqn:

```python
def equationroots( a, b, c):
    import math 
    dis = b * b - 4 * a * c 
    sqrt_val = math.sqrt(abs(dis)) 
    if dis > 0: 
        print(" real and different roots ") 
        print((-b + sqrt_val)/(2 * a)) 
        print((-b - sqrt_val)/(2 * a)) 
    elif dis == 0: 
        print(" real and same roots") 
        print(-b / (2 * a)) 
    else:
        print("Complex Roots") 
        print(- b / (2 * a), " + i", sqrt_val) 
        print(- b / (2 * a), " - i", sqrt_val)

print('Solution of quadratic equation for given values of a, b and c:')
equationroots(3,4,5)
```

### 4- Write different functions which takes lists. They should calculate_mean, calculate_median, calculate_mode, calculate_range, calculate_variance, calculate_std:

```python
A = [1,2,3,3,2,2,2,1]
```

```python
#4-1
#calculate_mean function:
def calculate_mean(a):
    mean = sum(a)/len(a)
    print('Mean:', mean)
calculate_mean(A)
```

```python
#4-2
#calculate_median function:
def calculate_median(sample):
        n = len(sample)
        index = n // 2
        if n % 2:
            return sorted(sample)[index]
        return sum(sorted(sample)[index - 1:index + 1]) / 2
calculate_median(A)
```

```python
#4-3
#calculate_mode function:
def calculate_mode(a):
    from collections import Counter
    c = Counter(a)
    mode = [k for k, v in c.items() if  v== c.most_common(1)[0][1]]
    print('Mode:', mode)
calculate_mode(A)
```

```python
#4-4
#calculate_range function
def calculate_range(a):
    Maximum = max(a)
    Minimum = min(a)
    range =  Maximum- Minimum
    print('Range: ', range)
calculate_range(A)
```

```python
#4-5
#calculate_variance function:
def calculate_variance(a):
    mu = sum(a)/len(a)
    element =[]
    for i in a:
        c= i-mu
        element.append(c*c)
    print('Variance:', sum(element)/(len(element)-1))
calculate_variance(A)
```

```python
#4-6
#calculate_std function:
def calculate_std(a):
    mu = sum(a)/len(a)
    element =[]
    for i in a:
        c= mu-i
        element.append(c*c)
    std  =(sum(element)/(len(element)-1))**(1/2)
    print('Standard Deviation:', std)
calculate_std(A)
```

### 1- Difference between map, filter and reduce:

  1) Map function-
     - Map function iterates through all items in the given iterable and execute the function we passed as an argument on each        of them.
     - We can pass many iterable objects as we want after passing the function.
     - syntax is - map(function,iterable(s))

   2) Filter function-
     - Filter() forms a new list that contains only elements that satisfy a certain contain condition.
     - Syntax is - filter(function,iterable(s))

   3) Reduce function- 
     - Reduce() function is different from map and filter .it does not return a new list based on the function.
     - Syntax is - reduce(function,sequence[, initial])
     - Reduce works by calling the function.

### 2- Difference between higher order function, closure and decorator:

   1) Higher order function-
     - A function is called higher order function if it contain other function as a parameter or returns a function as an              output i.e,the functions that operate with another function.
     - Properties of higher order function:
        1) A function is an instance of the Object type.
        2) You can store the function in a variable.
        3) You can pass the function as a parameter to another function.
        4) You can return the function from a function.
        5) You can store them in data structures such as hash tables, lists.

   2) Closure- 
    - A closure is a function object that remembers values in enclosing scopes even if they are not present in memory.
    - Closure allows the function to access those captured variable through the closure's copies of the values or                     references,even when the function is invoked outside their scope.

   3) Decorator- 
     - It allows user to wrap another function in order to extend the behavior of the wrapped function,without permanently            modifying it.

### 2.Create image mask with rom_jul.png and instantiate, generate and display a word cloud object image:

```python
import wordcloud
from wordcloud import WordCloud,STOPWORDS, ImageColorGenerator
from PIL import Image
import numpy as np
filename='rom_jul.png'
with Image.open(filename) as img:
    img.load()
img.show()
img_gray=img.convert('L')
threshold=125
img_threshold=img_gray.point(
    lambda x:255 if x>threshold else 0
)
mask_img=np.array(img_threshold)
comment_words=''
stopwords=set(STOPWORDS)
text=open('romeo_and_juliet.txt').read()
Text=str(text)
wc=WordCloud(width=3000,
                     height=3000,
                     stopwords=stopwords, 
                     max_words=200,
                     mask=mask_img,
                     min_font_size=10).generate(text)
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(16,20))
plt.imshow(wc,
           interpolation='bilinear')
plt.axis('off')
plt.show()
```
